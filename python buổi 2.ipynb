```

#Câu 6:
def doc_so(n):
    chu_so = {
        0: "không",
        1: "một",
        2: "hai",
        3: "ba",
        4: "bốn",
        5: "năm",
        6: "sáu",
        7: "bảy",
        8: "tám",
        9: "chín"
    }

    if n < 10:
        return chu_so[n].capitalize()

    chuc = n // 10
    don_vi = n % 10

    if chuc == 1:
        if don_vi == 0:
            return "Mười"
        elif don_vi == 5:
            return "Mười lăm"
        else:
            return "Mười " + chu_so[don_vi]
    else:
        phan_chuc = chu_so[chuc] + " mươi"
        if don_vi == 0:
            return phan_chuc.capitalize()
        elif don_vi == 1:
            return phan_chuc.capitalize() + " mốt"
        elif don_vi == 4:
            return phan_chuc.capitalize() + " tư"
        elif don_vi == 5:
            return phan_chuc.capitalize() + " lăm"
        else:
            return phan_chuc.capitalize() + " " + chu_so[don_vi]

# Nhập số
n = int(input("Nhập số (0-99): "))
if 0 <= n <= 99:
    print(doc_so(n))
else:
    print("Vui lòng nhập số từ 0 đến 99.")
```

```
#Câu 7:
def la_nam_nhuan(nam):
    # Năm nhuận nếu chia hết cho 400 hoặc chia hết cho 4 mà không chia hết cho 100
    return (nam % 400 == 0) or (nam % 4 == 0 and nam % 100 != 0)

def ngay_ke_sau(ngay, thang, nam):
    # Số ngày tối đa từng tháng
    if thang in [1, 3, 5, 7, 8, 10, 12]:
        ngay_max = 31
    elif thang in [4, 6, 9, 11]:
        ngay_max = 30
    elif thang == 2:
        ngay_max = 29 if la_nam_nhuan(nam) else 28
    else:
        return None  # Tháng không hợp lệ

    # Tính ngày kế tiếp
    if ngay < ngay_max:
        ngay += 1
    else:
        ngay = 1
        if thang == 12:
            thang = 1
            nam += 1
        else:
            thang += 1
    return ngay, thang, nam

# Nhập ngày tháng năm
ngay = int(input("Nhập ngày: "))
thang = int(input("Nhập tháng: "))
nam = int(input("Nhập năm: "))

ket_qua = ngay_ke_sau(ngay, thang, nam)

if ket_qua:
    n, t, y = ket_qua
    print(f"Ngày kế sau là: {n}/{t}/{y}")
else:
    print("Ngày tháng năm không hợp lệ.")
```

```
#Câu 10:
def giai_thua(k):
    # Tính k! (giai thừa)
    if k == 0 or k == 1:
        return 1
    result = 1
    for i in range(2, k+1):
        result *= i
    return result

def tinh_tong(x, n):
    tong = 0.0
    for i in range(1, n+1):
        tong += (x ** i) / giai_thua(i)
    return tong

# Nhập dữ liệu
x = float(input("Nhập giá trị x: "))
n = int(input("Nhập giá trị n: "))

ket_qua = tinh_tong(x, n)
print(f"Kết quả S({x}, {n}) = {ket_qua}")
```

Trong trường hợp người dùng nhập một số bằng với giá trị của m (là 100), điều kiện if n == m: cũng sẽ đúng và thoát vòng lặp.

```
#cau 18.1
def ve_tam_giac_rong(n):
    for i in range(n):
        for j in range(n):
            if j == 0 or i == n - 1 or i == j:
                print("*", end=" ")
            else:
                print(" ", end=" ")
        print()

# Ví dụ vẽ tam giác cao 5
ve_tam_giac_rong(5)
```

```
# cau 18.2
def ve_tam_giac_nguoc(n):
    for i in range(n, 0, -1): # Lặp từ n đến 1
        for j in range(i):
            print("*", end=" ")
        print()

# Ví dụ vẽ tam giác vuông ngược cao 5
ve_tam_giac_nguoc(5)
```

```
# cau 19:
import math

def tinh_S(x, n):
    """
    Tính giá trị của S(x,n) = x + x^3/3! + x^5/5! + ... + x^(2n+1)/(2n+1)!

    Args:
        x (float): Giá trị của biến x.
        n (int): Số hạng cuối cùng của chuỗi có bậc là 2n+1.

    Returns:
        float: Giá trị của S(x,n).
    """
    S = 0
    for i in range(n + 1):
        bac = 2 * i + 1
        mau_so = math.factorial(bac)
        so_hang = (x ** bac) / mau_so
        S += so_hang
    return S

# Nhập giá trị từ người dùng
try:
    x_nhap = float(input("Nhập giá trị cho x: "))
    n_nhap = int(input("Nhập giá trị cho n: "))

    if n_nhap < 0:
        print("Giá trị của n phải là số nguyên không âm.")
    else:
        ket_qua = tinh_S(x_nhap, n_nhap)
        print(f"Giá trị của S({x_nhap}, {n_nhap}) là: {ket_qua}")

except ValueError:
    print("Đầu vào không hợp lệ. Vui lòng nhập số cho x và số nguyên cho n.")
```
