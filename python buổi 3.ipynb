#Câu 6: Những giá trị nào có thể xuất hiện trong randrange(0, 100)

Giá trị có thể xuất hiện trong hàm randrange(0,100) là bất kỳ số nguyên nào từ 0 đến 99. Cụ thể, hàm này sẽ chọn ngẫu nhiên một số nguyên N thỏa mãn điều kiện 0≤N<100. Điều này có nghĩa là 0 là giá trị nhỏ nhất có thể xuất hiện, và 99 là giá trị lớn nhất có thể xuất hiện (vì giá trị 100 là giới hạn trên và không được bao gồm).

```
#Câu 7: Tính và xuất độ dài đoạn AB
import math # Import module math để sử dụng hàm sqrt (căn bậc hai)

# 1. Nhập toạ độ điểm A
print("--- Nhập toạ độ điểm A ---")
xA = float(input("Nhập hoành độ xA: "))
yA = float(input("Nhập tung độ yA: "))

# 2. Nhập toạ độ điểm B
print("\n--- Nhập toạ độ điểm B ---")
xB = float(input("Nhập hoành độ xB: "))
yB = float(input("Nhập tung độ yB: "))

# 3. Tính toán độ dài đoạn AB
# Công thức: AB = sqrt((xB - xA)^2 + (yB - yA)^2)

# Tính hiệu số hoành độ và tung độ
delta_x = xB - xA
delta_y = yB - yA

# Tính tổng bình phương các hiệu số
tong_binh_phuong = delta_x**2 + delta_y**2

# Tính căn bậc hai để tìm độ dài AB
do_dai_AB = math.sqrt(tong_binh_phuong)

# 4. Xuất kết quả
print("\n--- Kết quả ---")
print(f"Toạ độ điểm A: ({xA}, {yA})")
print(f"Toạ độ điểm B: ({xB}, {yB})")
print(f"Độ dài đoạn AB là: {do_dai_AB}")
```

```
#Câu 8: Viết chương trình tính loga
import math # Cần import module math để sử dụng hàm log

# 1. Nhập giá trị x và cơ số a
print("--- Tính Logarit cơ số a của x (loga x) ---")

# Nhập x (số lấy logarit)
# Điều kiện: x > 0
try:
    x = float(input("Nhập giá trị x (x > 0): "))
    if x <= 0:
        print("Lỗi: x phải là số dương.")
        exit()
except ValueError:
    print("Lỗi: Giá trị x không hợp lệ.")
    exit()

# Nhập a (cơ số)
# Điều kiện: a > 0 và a != 1
try:
    a = float(input("Nhập cơ số a (a > 0 và a != 1): "))
    if a <= 0 or a == 1:
        print("Lỗi: Cơ số a phải lớn hơn 0 và khác 1.")
        exit()
except ValueError:
    print("Lỗi: Giá trị a không hợp lệ.")
    exit()

# 2. Tính toán loga x
# Sử dụng hàm math.log(x, a)
gia_tri_loga_x = math.log(x, a)

# 3. Xuất kết quả
print("\n--- Kết quả ---")
print(f"logarit cơ số {a} của {x} là: {gia_tri_loga_x}")
```

```
#Câu 9: Viết chương trình tính căn bậc 2 lồng nhau

import math

def tinh_can_long_nhau(n):
    """
    Tính giá trị của biểu thức S(n) = sqrt(2 + sqrt(2 + ... + sqrt(2)))
    với n dấu căn.
    """
    if n <= 0:
        return 0.0

    # Khởi tạo giá trị ban đầu là căn bậc hai cuối cùng: sqrt(2)
    # Đây là giá trị của S(1)
    S = math.sqrt(2)

    # Vòng lặp tính toán từ trong ra ngoài (hoặc từ S(2) đến S(n))
    # Bắt đầu từ i = 2 vì S(1) đã được tính ở trên
    for i in range(2, n + 1):
        # Công thức lặp: S(i) = sqrt(2 + S(i-1))
        # Trong đó, S là giá trị S(i-1) từ vòng lặp trước
        S = math.sqrt(2 + S)

    return S

# 1. Nhập giá trị n
print("--- Tính tổng S(n) với n dấu căn lồng nhau ---")

try:
    n = int(input("Nhập số lượng dấu căn n (n >= 1): "))
    if n <= 0:
        print("Lỗi: n phải là số nguyên dương (n >= 1).")
    else:
        # 2. Tính toán và xuất kết quả
        ket_qua = tinh_can_long_nhau(n)
        print(f"\nGiá trị của S({n}) với {n} dấu căn là: {ket_qua}")

except ValueError:
    print("Lỗi: Đầu vào không hợp lệ. Vui lòng nhập một số nguyên.")
```

```
#Câu 12: Hàm oscillate

def oscillate(start, count):
    """
    Hàm generator tạo ra chuỗi số theo mẫu:
    Tăng dần từ start đến 0, sau đó 0, rồi giảm dần.

    Tham số:
    - start: Giá trị khởi đầu (ví dụ: -3)
    - count: Số lần lặp cho phần giảm dần sau 0 (ví dụ: 5)
    """

    # 1. Phần Tăng dần (Từ 'start' đến 0)
    current = start
    while current < 0:
        yield current
        current += 1

    # 2. Phần Giữ giá trị 0
    yield 0
    yield 0

    # 3. Phần Giảm dần (Từ 0, giảm count lần)
    # count = 5 sẽ tạo ra 0 - 1 = -1, ..., 0 - 4 = -4
    # Ta đã in 0, nên bắt đầu từ -1
    for i in range(1, count):  # Chạy 4 lần: i=1, 2, 3, 4
        yield -i # Tạo ra -1, -2, -3, -4

    # 4. Phần Lặp giá trị cuối (-4)
    # Lấy giá trị cuối cùng được tạo ra: -(count - 1) = -4
    yield -(count - 1)

# Kiểm tra kết quả
# for n in oscillate(-3, 5):
#     print(n, end=' ')
# print()
# Output: -3 -2 -1 0 0 -1 -2 -3 -4 -4
```

```
#Câu 13: Hàm kiểm tra số hoàn thiện, số thịnh vượng
def sum_proper_divisors(n):
    """
    Tính tổng các ước số dương của số nguyên n, không kể n.
    Ưu tiên tìm ước số đến căn bậc hai của n để tối ưu tốc độ.
    """
    if n <= 0:
        return 0
    if n == 1:
        # Ước số dương của 1 chỉ là 1. Tổng các ước số không kể nó là 0.
        return 0

    total_sum = 1 # Bắt đầu với 1, vì 1 luôn là ước số của mọi n > 1
    limit = int(n**0.5) # Giới hạn tìm ước số là căn bậc hai của n

    for i in range(2, limit + 1):
        if n % i == 0:
            # i là ước số
            total_sum += i
            # n // i cũng là ước số (không phải i)
            j = n // i
            if i != j:
                total_sum += j

    return total_sum
```
